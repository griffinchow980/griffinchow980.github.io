**ğŸ”„ R (Recency) - æœ€è¿‘æ¶ˆè´¹æ—¶é—´**
- **å«ä¹‰**ï¼šç”¨æˆ·æœ€è¿‘ä¸€æ¬¡è´­ä¹°è·ä»Šçš„æ—¶é—´
- **è§£è¯»**ï¼šRå€¼è¶Šå°ï¼Œè¡¨ç¤ºç”¨æˆ·è¶Šæ´»è·ƒï¼Œæµå¤±é£é™©è¶Šä½

- **è§£è¯»**ï¼šMå€¼è¶Šå¤§ï¼Œè¡¨ç¤ºç”¨æˆ·è´¡çŒ®åº¦è¶Šé«˜

### 7.1.3 æ„å»ºRFMæ•°æ®é›†

```python
# æ­¥éª¤1ï¼šæå–RFMåŸå§‹æ•°æ®
rfm_df = df[['User_ID', 'Last_Login_Days_Ago', 'Purchase_Frequency', 'Total_Spending']].copy()

# é‡å‘½ååˆ—ä¸ºæ ‡å‡†çš„RFMå‘½å
rfm_df = rfm_df.rename(columns={
    'Last_Login_Days_Ago': 'R',  # Recency - æœ€è¿‘ç™»å½•å¤©æ•°
    'Purchase_Frequency': 'F',    # Frequency - è´­ä¹°é¢‘ç‡  
    'Total_Spending': 'M'         # Monetary - æ€»æ¶ˆè´¹é‡‘é¢
})

print("="*70)
print("ğŸ“Š RFMåŸå§‹æ•°æ®é¢„è§ˆ")
print("="*70)
display(rfm_df.head(10))

print("\n" + "="*70)
print("ğŸ“ˆ RFMæŒ‡æ ‡æè¿°æ€§ç»Ÿè®¡")
print("="*70)
display(rfm_df[['R', 'F', 'M']].describe().round(2))
```

### 7.1.4 RFMåˆ†æ•°è®¡ç®—

**åˆ†ç®±æ–¹æ³•é€‰æ‹©ï¼šç™¾åˆ†ä½æ•°åˆ†ç®±æ³•**

å¸¸ç”¨çš„RFMæ‰“åˆ†æ–¹æ³•åŒ…æ‹¬ï¼š
1. **ç­‰è·åˆ†ç®±æ³•**ï¼šæŒ‰æ•°å€¼èŒƒå›´ç­‰è·åˆ’åˆ†
2. **ç™¾åˆ†ä½æ•°åˆ†ç®±æ³•**ï¼šæŒ‰æ•°æ®ç™¾åˆ†ä½æ•°åˆ’åˆ†ï¼ˆæœ¬é¡¹ç›®é‡‡ç”¨ï¼‰
3. **åŠ æƒæ±‡æ€»æ³•**ï¼šèµ‹äºˆä¸åŒæƒé‡åæ±‡æ€»
4. **æ ‡å‡†åŒ–æ‰“åˆ†æ³•**ï¼šæ ‡å‡†åŒ–åæ˜ å°„åˆ°åˆ†æ•°åŒºé—´

**ä¸ºä»€ä¹ˆé€‰æ‹©ç™¾åˆ†ä½æ•°åˆ†ç®±æ³•ï¼Ÿ**
- âœ… ä¸å—å¼‚å¸¸å€¼å½±å“ï¼Œæ›´ç¨³å¥
- âœ… èƒ½å¤Ÿä¿è¯æ¯ä¸ªåˆ†æ•°æ®µç”¨æˆ·æ•°é‡ç›¸å¯¹å‡è¡¡
- âœ… å®æ–½ç®€ä¾¿ï¼Œç»“æœç›´è§‚
- âœ… é€‚åˆæ•°æ®åˆ†å¸ƒä¸å‡åŒ€çš„åœºæ™¯

**å…·ä½“å®ç°ï¼š**
- ä½¿ç”¨ `pd.qcut()` å°†æ•°æ®åˆ†ä¸º5ä¸ªç­‰çº§ï¼ˆ1-5åˆ†ï¼‰
- Rå€¼éœ€è¦åè½¬ï¼ˆå¤©æ•°è¶Šå°åˆ†æ•°è¶Šé«˜ï¼‰
- Få’ŒMå€¼æ­£å‘æ‰“åˆ†ï¼ˆæ•°å€¼è¶Šå¤§åˆ†æ•°è¶Šé«˜ï¼‰

```python
# æ­¥éª¤2ï¼šè®¡ç®—RFMåˆ†æ•°ï¼ˆ1-5åˆ†ï¼‰

# Råˆ†æ•°ï¼šå¤©æ•°è¶Šå°ï¼Œåˆ†æ•°è¶Šé«˜ï¼ˆä½¿ç”¨è´Ÿå€¼åè½¬ï¼‰
rfm_df['R_Score'] = pd.qcut(
    -rfm_df['R'],           # ä½¿ç”¨è´Ÿå€¼åè½¬ï¼Œä½¿å¾—å¤©æ•°å°çš„ç”¨æˆ·å¾—é«˜åˆ†
    q=5,                    # åˆ†ä¸º5ä¸ªç­‰çº§
    labels=False,           # è¿”å›æ•°å€¼è€ŒéåŒºé—´æ ‡ç­¾
    duplicates='drop'       # å¤„ç†é‡å¤è¾¹ç•Œå€¼
) + 1                       # ä»1å¼€å§‹è®¡åˆ†ï¼ˆè€Œé0ï¼‰

# Fåˆ†æ•°ï¼šé¢‘ç‡è¶Šé«˜ï¼Œåˆ†æ•°è¶Šé«˜
rfm_df['F_Score'] = pd.qcut(
    rfm_df['F'],
    q=5,
    labels=False,
    duplicates='drop'
) + 1

# Måˆ†æ•°ï¼šæ¶ˆè´¹é‡‘é¢è¶Šé«˜ï¼Œåˆ†æ•°è¶Šé«˜
rfm_df['M_Score'] = pd.qcut(
    rfm_df['M'],
    q=5,
    labels=False,
    duplicates='drop'
) + 1

print("="*70)
print("âœ… RFMåˆ†æ•°è®¡ç®—å®Œæˆ")
print("="*70)
display(rfm_df.head(10))

# åˆ†æ•°åˆ†å¸ƒç»Ÿè®¡
print("\n" + "="*70)
print("ğŸ“Š RFMåˆ†æ•°åˆ†å¸ƒç»Ÿè®¡")
print("="*70)
print("\nRåˆ†æ•°åˆ†å¸ƒ:")
print(rfm_df['R_Score'].value_counts().sort_index())
print("\nFåˆ†æ•°åˆ†å¸ƒ:")
print(rfm_df['F_Score'].value_counts().sort_index())
print("\nMåˆ†æ•°åˆ†å¸ƒ:")
print(rfm_df['M_Score'].value_counts().sort_index())
```

### 7.1.5 ç”¨æˆ·æ ‡ç­¾åˆ†ç±»

åŸºäºRFMåˆ†æ•°ï¼Œæˆ‘ä»¬éœ€è¦è®¾è®¡ä¸€ä¸ªåˆ†ç±»å‡½æ•°ï¼Œä¸ºæ¯ä¸ªç”¨æˆ·æ‰“ä¸Šå¯¹åº”çš„æ ‡ç­¾ã€‚

```python
# æ­¥éª¤3ï¼šè®¾è®¡RFMæ ‡ç­¾åˆ†ç±»å‡½æ•°

def rfm_label(row):
    """
    æ ¹æ®RFMåˆ†æ•°ä¸ºç”¨æˆ·æ‰“æ ‡ç­¾
    
    åˆ†ç±»é€»è¾‘ï¼š
    - R >= 4: æœ€è¿‘æ´»è·ƒï¼ˆRé«˜ï¼‰
    - R < 4: è¾ƒä¹…æœªç™»å½•ï¼ˆRä½ï¼‰
    - F >= 4: é«˜é¢‘è´­ä¹°ï¼ˆFé«˜ï¼‰
    - F < 4: ä½é¢‘è´­ä¹°ï¼ˆFä½ï¼‰
    - M >= 4: é«˜æ¶ˆè´¹ï¼ˆMé«˜ï¼‰
    - M < 4: ä½æ¶ˆè´¹ï¼ˆMä½ï¼‰
    """
    R, F, M = row['R_Score'], row['F_Score'], row['M_Score']
    
    # é«˜æ¶ˆè´¹ç”¨æˆ·ï¼ˆM >= 4ï¼‰- é‡è¦å®¢æˆ·
    if M >= 4:
        if R >= 4 and F >= 4:
            return 'é‡è¦ä»·å€¼å®¢æˆ·'     # æœ€è¿‘æ´»è·ƒ + é«˜é¢‘ + é«˜æ¶ˆè´¹ = VIP
        elif R < 4 and F >= 4:
            return 'é‡è¦ä¿æŒå®¢æˆ·'     # ä¹…æœªç™»å½• + é«˜é¢‘ + é«˜æ¶ˆè´¹ = éœ€å”¤å›
        elif R >= 4 and F < 4:
            return 'é‡è¦å‘å±•å®¢æˆ·'     # æœ€è¿‘æ´»è·ƒ + ä½é¢‘ + é«˜æ¶ˆè´¹ = æ½œåŠ›è‚¡
        else:  # R < 4 and F < 4
            return 'é‡è¦æŒ½ç•™å®¢æˆ·'     # ä¹…æœªç™»å½• + ä½é¢‘ + é«˜æ¶ˆè´¹ = æµå¤±é£é™©
    
    # ä½æ¶ˆè´¹ç”¨æˆ·ï¼ˆM < 4ï¼‰- ä¸€èˆ¬å®¢æˆ·
    else:
        if R >= 4 and F >= 4:
            return 'ä¸€èˆ¬ä»·å€¼å®¢æˆ·'     # æœ€è¿‘æ´»è·ƒ + é«˜é¢‘ + ä½æ¶ˆè´¹ = å¯æå‡å®¢å•ä»·
        elif R < 4 and F >= 4:
            return 'ä¸€èˆ¬ä¿æŒå®¢æˆ·'     # ä¹…æœªç™»å½• + é«˜é¢‘ + ä½æ¶ˆè´¹ = ç»´æŠ¤å¿ è¯šåº¦
        elif R >= 4 and F < 4:
            return 'ä¸€èˆ¬å‘å±•å®¢æˆ·'     # æœ€è¿‘æ´»è·ƒ + ä½é¢‘ + ä½æ¶ˆè´¹ = æ–°ç”¨æˆ·
        else:  # R < 4 and F < 4
            return 'ä¸€èˆ¬æŒ½ç•™å®¢æˆ·'     # ä¹…æœªç™»å½• + ä½é¢‘ + ä½æ¶ˆè´¹ = å·²æµå¤±

# åº”ç”¨æ ‡ç­¾å‡½æ•°
rfm_df['RFM_Label'] = rfm_df.apply(rfm_label, axis=1)

print("="*70)
print("âœ… ç”¨æˆ·æ ‡ç­¾åˆ†ç±»å®Œæˆ")
print("="*70)
display(rfm_df.head(10))

# ç»Ÿè®¡å„ç±»ç”¨æˆ·æ•°é‡
print("\n" + "="*70)
print("ğŸ“Š ç”¨æˆ·åˆ†ç±»ç»Ÿè®¡")
print("="*70)
label_counts = rfm_df['RFM_Label'].value_counts()
for label, count in label_counts.items():
    percentage = count / len(rfm_df) * 100
    print(f"{label}: {count} äºº ({percentage:.1f}%)")
```

### 7.1.7 RFMåˆ†æç»“è®ºä¸è¥é”€ç­–ç•¥

## 7.2 K-Meansèšç±»åˆ†æ

è™½ç„¶RFMæ¨¡å‹æä¾›äº†æ˜ç¡®çš„ç”¨æˆ·åˆ†ç±»ï¼Œä½†å®ƒåŸºäºäººä¸ºè®¾å®šçš„é˜ˆå€¼ï¼Œå…·æœ‰ä¸€å®šä¸»è§‚æ€§ã€‚ä¸ºäº†è·å¾—æ›´å®¢è§‚ã€æ•°æ®é©±åŠ¨çš„ç”¨æˆ·åˆ†ç¾¤ç»“æœï¼Œæˆ‘ä»¬å°†é‡‡ç”¨K-Meansèšç±»ç®—æ³•è¿›è¡Œæ·±åº¦åˆ†æã€‚

### 7.2.1 ä¸ºä»€ä¹ˆéœ€è¦K-Meansèšç±»ï¼Ÿ

**RFMæ¨¡å‹çš„å±€é™æ€§ï¼š**
- âŒ ä¾èµ–äººä¸ºè®¾å®šçš„é˜ˆå€¼ï¼ˆå¦‚Râ‰¥4ä¸ºé«˜ï¼‰
- âŒ å¯èƒ½æ— æ³•æ•æ‰å¤æ‚çš„ç”¨æˆ·è¡Œä¸ºæ¨¡å¼
- âŒ å¿½ç•¥äº†å…¶ä»–é‡è¦çš„ç”¨æˆ·ç‰¹å¾

**K-Meansèšç±»çš„ä¼˜åŠ¿ï¼š**
- âœ… å®Œå…¨æ•°æ®é©±åŠ¨ï¼Œæ— äººä¸ºåè§
- âœ… å¯ä»¥æ•´åˆå¤šç»´åº¦ç‰¹å¾è¿›è¡Œç»¼åˆåˆ†æ
- âœ… è‡ªåŠ¨å‘ç°æ•°æ®ä¸­çš„è‡ªç„¶åˆ†ç¾¤
- âœ… æ›´çµæ´»çš„ç‰¹å¾é€‰æ‹©å’Œç»„åˆ

### 7.2.2 ç‰¹å¾å·¥ç¨‹ç­–ç•¥

ç»è¿‡å¤šæ¬¡å®éªŒï¼Œæˆ‘ä»¬é‡‡ç”¨ä»¥ä¸‹ä¼˜åŒ–ç­–ç•¥ï¼š

**æ–¹æ¡ˆä¸€ï¼šä»…ä½¿ç”¨RFMåˆ†æ•°** âŒ
- æ•ˆæœï¼šèšç±»æ•ˆæœä¸ç†æƒ³ï¼Œåˆ†ç¾¤è¾¹ç•Œæ¨¡ç³Š
- åŸå› ï¼šç‰¹å¾ç»´åº¦å°‘ï¼Œä¿¡æ¯é‡ä¸è¶³

**æ–¹æ¡ˆäºŒï¼šä½¿ç”¨åŸå§‹RFMå€¼** âŒ  
- æ•ˆæœï¼šå—å¼‚å¸¸å€¼å½±å“å¤§ï¼Œèšç±»ä¸ç¨³å®š
- åŸå› ï¼šæ•°æ®åˆ†å¸ƒä¸å‡ï¼Œé‡çº²å·®å¼‚å¤§

**æ–¹æ¡ˆä¸‰ï¼šç»¼åˆç‰¹å¾å·¥ç¨‹** âœ… **ï¼ˆæœ€ä¼˜æ–¹æ¡ˆï¼‰**
- æ•´åˆRFMæ ¸å¿ƒæŒ‡æ ‡ + ç”¨æˆ·è¡Œä¸ºç‰¹å¾ + ç”¨æˆ·å±æ€§
- ä½¿ç”¨RobustScalerå¤„ç†å¼‚å¸¸å€¼
- ç‰¹å¾æ ‡å‡†åŒ–ä¿è¯å…¬å¹³æ€§

### 7.2.3 ä¼˜åŒ–çš„ç‰¹å¾å·¥ç¨‹

```python

# æ–¹æ¡ˆé€‰æ‹©ï¼šç»¼åˆå¤šç»´åº¦ç‰¹å¾
# æ•´åˆï¼šå¹´é¾„ã€æ”¶å…¥ã€RFMæŒ‡æ ‡ã€è¡Œä¸ºç‰¹å¾

# æ­¥éª¤1ï¼šæ„å»ºç‰¹å¾é›†
cluster_features = df[[
    # ç”¨æˆ·å±æ€§
    'Age',
    'Income',
    # RFMæ ¸å¿ƒæŒ‡æ ‡
    'Last_Login_Days_Ago',      # R - æœ€è¿‘ç™»å½•
    'Purchase_Frequency',        # F - è´­ä¹°é¢‘ç‡
    'Total_Spending',            # M - æ€»æ¶ˆè´¹
    # è¡Œä¸ºç‰¹å¾
    'Average_Order_Value',       # å¹³å‡è®¢å•ä»·å€¼
    'Time_Spent_on_Site_Minutes',  # åœç•™æ—¶é•¿
    'Pages_Viewed',              # æµè§ˆé¡µæ•°
]].copy()

# æ­¥éª¤2ï¼šç‰¹å¾å·¥ç¨‹ - åˆ›å»ºè¡ç”Ÿç‰¹å¾

# ç”¨æˆ·ä»·å€¼å¯†åº¦ï¼ˆæ¶ˆè´¹é‡‘é¢/è´­ä¹°æ¬¡æ•°ï¼‰- åæ˜ å®¢å•ä»·å€¾å‘
cluster_features['Value_Density'] = cluster_features['Total_Spending'] / (cluster_features['Purchase_Frequency'] + 1)

# æ´»è·ƒåº¦æŒ‡æ ‡ï¼ˆè¶Šå°è¶Šæ´»è·ƒï¼‰
cluster_features['Activity_Score'] = 30 - cluster_features['Last_Login_Days_Ago']

# æµè§ˆæ•ˆç‡ï¼ˆæµè§ˆé¡µæ•°/åœç•™æ—¶é•¿ï¼‰
cluster_features['Browse_Efficiency'] = cluster_features['Pages_Viewed'] / (cluster_features['Time_Spent_on_Site_Minutes'] + 1)

print("="*70)
print("ğŸ“Š èšç±»ç‰¹å¾é›†æ„å»ºå®Œæˆ")
print("="*70)
print(f"ç‰¹å¾æ•°é‡: {cluster_features.shape[1]}")
print(f"æ ·æœ¬æ•°é‡: {cluster_features.shape[0]}")
print("\nç‰¹å¾åˆ—è¡¨:")
for i, col in enumerate(cluster_features.columns, 1):
    print(f"  {i}. {col}")

display(cluster_features.head(10))
display(cluster_features.describe().round(2))
```

### 7.2.4 æ•°æ®æ ‡å‡†åŒ–

**ä¸ºä»€ä¹ˆéœ€è¦æ ‡å‡†åŒ–ï¼Ÿ**
- ä¸åŒç‰¹å¾çš„é‡çº²å·®å¼‚å·¨å¤§ï¼ˆå¦‚Ageåœ¨18-64ï¼ŒTotal_Spendingåœ¨112-4999ï¼‰
- æœªæ ‡å‡†åŒ–ä¼šå¯¼è‡´å¤§æ•°å€¼ç‰¹å¾ä¸»å¯¼èšç±»ç»“æœ
- æ ‡å‡†åŒ–ä¿è¯æ‰€æœ‰ç‰¹å¾åœ¨èšç±»ä¸­æƒé‡å¹³ç­‰

**æ ‡å‡†åŒ–æ–¹æ³•é€‰æ‹©ï¼šRobustScaler**
- âœ… åŸºäºä¸­ä½æ•°å’Œå››åˆ†ä½æ•°ï¼Œå¯¹å¼‚å¸¸å€¼æ›´ç¨³å¥
- âœ… é€‚åˆæœ‰ååˆ†å¸ƒçš„æ•°æ®
- âœ… æ¯”StandardScaleræ›´é€‚åˆå®é™…ä¸šåŠ¡æ•°æ®

```python
# ä½¿ç”¨RobustScalerè¿›è¡Œæ ‡å‡†åŒ–
robust_scaler = RobustScaler()
scaled_features = robust_scaler.fit_transform(cluster_features)

# åˆ›å»ºæ ‡å‡†åŒ–åçš„DataFrameï¼ˆä¾¿äºåç»­åˆ†æï¼‰
scaled_df = pd.DataFrame(
    scaled_features,
    columns=cluster_features.columns,
    index=cluster_features.index
)

print("="*70)
print("âœ… æ•°æ®æ ‡å‡†åŒ–å®Œæˆ")
print("="*70)
print(f"æ ‡å‡†åŒ–æ–¹æ³•: RobustScaler")
print(f"æ ‡å‡†åŒ–åæ•°æ®å½¢çŠ¶: {scaled_features.shape}")

print("\næ ‡å‡†åŒ–å‰åå¯¹æ¯”ï¼ˆå‰5ä¸ªæ ·æœ¬ï¼‰:")
print("\nã€æ ‡å‡†åŒ–å‰ã€‘")
display(cluster_features.head())
print("\nã€æ ‡å‡†åŒ–åã€‘")
display(scaled_df.head())

# éªŒè¯æ ‡å‡†åŒ–æ•ˆæœ
print("\næ ‡å‡†åŒ–åçš„ç»Ÿè®¡ç‰¹å¾:")
display(scaled_df.describe().round(2))
```

### 7.2.5 ç¡®å®šæœ€ä¼˜Kå€¼ï¼ˆèšç±»æ•°é‡ï¼‰

è¿™æ˜¯èšç±»åˆ†ææœ€å…³é”®çš„æ­¥éª¤ï¼æˆ‘ä»¬å°†ä½¿ç”¨å¤šç§è¯„ä¼°æŒ‡æ ‡ç»¼åˆåˆ¤æ–­æœ€ä¼˜èšç±»æ•°ï¼š

**è¯„ä¼°æŒ‡æ ‡è¯´æ˜ï¼š**

1. **æ‰‹è‚˜æ³•ï¼ˆElbow Methodï¼‰** - åŸºäºæƒ¯æ€§ï¼ˆInertiaï¼‰
   - è§‚å¯Ÿæƒ¯æ€§éšKå€¼ä¸‹é™çš„æ›²çº¿
   - å¯»æ‰¾"è‚˜éƒ¨"æ‹ç‚¹ï¼ˆä¸‹é™é€Ÿåº¦æ˜æ˜¾å˜ç¼“å¤„ï¼‰
   - æƒ¯æ€§è¶Šå°è¡¨ç¤ºç°‡å†…ç´§å¯†åº¦è¶Šé«˜

2. **è½®å»“ç³»æ•°ï¼ˆSilhouette Scoreï¼‰**
   - èŒƒå›´ï¼š[-1, 1]
   - å€¼è¶Šæ¥è¿‘1è¡¨ç¤ºèšç±»æ•ˆæœè¶Šå¥½
   - ç»¼åˆè€ƒè™‘ç°‡å†…ç´§å¯†åº¦å’Œç°‡é—´åˆ†ç¦»åº¦

3. **Davies-BouldinæŒ‡æ•°ï¼ˆDB Indexï¼‰**
   - å€¼è¶Šå°è¡¨ç¤ºèšç±»æ•ˆæœè¶Šå¥½
   - è¡¡é‡ç°‡å†…ç›¸ä¼¼åº¦ä¸ç°‡é—´å·®å¼‚åº¦çš„æ¯”å€¼

4. **Calinski-HarabaszæŒ‡æ•°ï¼ˆCH Indexï¼‰**
   - å€¼è¶Šå¤§è¡¨ç¤ºèšç±»æ•ˆæœè¶Šå¥½
   - ç°‡é—´æ–¹å·®ä¸ç°‡å†…æ–¹å·®çš„æ¯”å€¼

```python
# è¯„ä¼°ä¸åŒKå€¼ï¼ˆ2-10ï¼‰çš„èšç±»æ•ˆæœ
k_range = range(2, 11)
metrics_results = {
    'K': [],
    'Inertia': [],
    'Silhouette': [],
    'Davies_Bouldin': [],
    'Calinski_Harabasz': []
}

print("="*70)
print("ğŸ”„ æ­£åœ¨è¯„ä¼°ä¸åŒKå€¼çš„èšç±»æ•ˆæœ...")
print("="*70)

for k in k_range:
    # è®­ç»ƒK-Meansæ¨¡å‹
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10, max_iter=300)
    labels = kmeans.fit_predict(scaled_features)
    
    # è®¡ç®—å„é¡¹è¯„ä¼°æŒ‡æ ‡
    inertia = kmeans.inertia_
    silhouette = silhouette_score(scaled_features, labels)
    db_score = davies_bouldin_score(scaled_features, labels)
    ch_score = calinski_harabasz_score(scaled_features, labels)
    
    # å­˜å‚¨ç»“æœ
    metrics_results['K'].append(k)
    metrics_results['Inertia'].append(inertia)
    metrics_results['Silhouette'].append(silhouette)
    metrics_results['Davies_Bouldin'].append(db_score)
    metrics_results['Calinski_Harabasz'].append(ch_score)
    
    print(f"K={k}: Inertia={inertia:.2f}, Silhouette={silhouette:.4f}, DB={db_score:.4f}, CH={ch_score:.2f}")

print("\nâœ… è¯„ä¼°å®Œæˆï¼")

# åˆ›å»ºæŒ‡æ ‡DataFrame
metrics_df = pd.DataFrame(metrics_results)
display(metrics_df)
```

### 7.2.6 æœ€ä¼˜Kå€¼åˆ†æä¸é€‰æ‹©


```python
# ç»¼åˆåˆ†ææœ€ä¼˜Kå€¼
print("="*70)
print("ğŸ“Š ç»¼åˆè¯„ä¼°åˆ†æ")
print("="*70)

# æ‰¾å‡ºå„æŒ‡æ ‡çš„æœ€ä¼˜Kå€¼
best_k_silhouette = metrics_df.loc[metrics_df['Silhouette'].idxmax(), 'K']
best_k_db = metrics_df.loc[metrics_df['Davies_Bouldin'].idxmin(), 'K']
best_k_ch = metrics_df.loc[metrics_df['Calinski_Harabasz'].idxmax(), 'K']

print(f"\nå„æŒ‡æ ‡æ¨èçš„æœ€ä¼˜Kå€¼ï¼š")
print(f"  â€¢ Silhouette Score æœ€å¤§å€¼: K = {best_k_silhouette} (åˆ†æ•°: {metrics_df.loc[metrics_df['K']==best_k_silhouette, 'Silhouette'].values[0]:.4f})")
print(f"  â€¢ Davies-Bouldinæœ€å°å€¼: K = {best_k_db} (åˆ†æ•°: {metrics_df.loc[metrics_df['K']==best_k_db, 'Davies_Bouldin'].values[0]:.4f})")
print(f"  â€¢ Calinski-Harabaszæœ€å¤§å€¼: K = {best_k_ch} (åˆ†æ•°: {metrics_df.loc[metrics_df['K']==best_k_ch, 'Calinski_Harabasz'].values[0]:.2f})")

# æ‰‹è‚˜æ³•åˆ†æï¼ˆè®¡ç®—ä¸‹é™ç‡ï¼‰
print(f"\næ‰‹è‚˜æ³•åˆ†æï¼ˆInertiaä¸‹é™ç‡ï¼‰ï¼š")
for i in range(1, len(metrics_df)):
    k_curr = metrics_df.loc[i, 'K']
    k_prev = metrics_df.loc[i-1, 'K']
    inertia_curr = metrics_df.loc[i, 'Inertia']
    inertia_prev = metrics_df.loc[i-1, 'Inertia']
    decrease_rate = (inertia_prev - inertia_curr) / inertia_prev * 100
    print(f"  K={int(k_prev)}â†’{int(k_curr)}: ä¸‹é™ {decrease_rate:.2f}%")

# ç»¼åˆæ¨è
print("\n" + "="*70)
print("ğŸ¯ ç»¼åˆæ¨è")
print("="*70)

# æ ‡å‡†åŒ–å„æŒ‡æ ‡å¹¶åŠ æƒæ‰“åˆ†
from sklearn.preprocessing import MinMaxScaler
scoring_scaler = MinMaxScaler()

# å¯¹éœ€è¦æœ€å¤§åŒ–çš„æŒ‡æ ‡æ­£å‘æ ‡å‡†åŒ–ï¼Œéœ€è¦æœ€å°åŒ–çš„æŒ‡æ ‡åå‘æ ‡å‡†åŒ–
silhouette_norm = scoring_scaler.fit_transform(metrics_df[['Silhouette']].values).flatten()
db_norm = 1 - scoring_scaler.fit_transform(metrics_df[['Davies_Bouldin']].values).flatten()  # åå‘
ch_norm = scoring_scaler.fit_transform(metrics_df[['Calinski_Harabasz']].values).flatten()

# ç»¼åˆå¾—åˆ†ï¼ˆå¯è°ƒæ•´æƒé‡ï¼‰
weights = {'silhouette': 0.4, 'db': 0.3, 'ch': 0.3}
ç»¼åˆå¾—åˆ† = (weights['silhouette'] * silhouette_norm + 
           weights['db'] * db_norm + 
           weights['ch'] * ch_norm)

metrics_df['ç»¼åˆå¾—åˆ†'] = ç»¼åˆå¾—åˆ†

# æ‰¾å‡ºç»¼åˆå¾—åˆ†æœ€é«˜çš„K
optimal_k = metrics_df.loc[metrics_df['ç»¼åˆå¾—åˆ†'].idxmax(), 'K']
optimal_score = metrics_df.loc[metrics_df['ç»¼åˆå¾—åˆ†'].idxmax(), 'ç»¼åˆå¾—åˆ†']

print(f"\nğŸ“ˆ å„Kå€¼ç»¼åˆå¾—åˆ†æ’åï¼š")
display(metrics_df[['K', 'Silhouette', 'Davies_Bouldin', 'Calinski_Harabasz', 'ç»¼åˆå¾—åˆ†']].sort_values('ç»¼åˆå¾—åˆ†', ascending=False))

print(f"\nâœ… æ¨èæœ€ä¼˜Kå€¼: {int(optimal_k)} (ç»¼åˆå¾—åˆ†: {optimal_score:.4f})")
print(f"\nç†ç”±ï¼š")
print(f"  1. è¯¥Kå€¼åœ¨å¤šä¸ªè¯„ä¼°æŒ‡æ ‡ä¸Šè¡¨ç°ä¼˜å¼‚")
print(f"  2. èšç±»æ•°é‡é€‚ä¸­ï¼Œä¾¿äºä¸šåŠ¡ç†è§£å’Œåº”ç”¨")
print(f"  3. ç»¼åˆæƒè¡¡äº†ç°‡å†…ç´§å¯†åº¦å’Œç°‡é—´åˆ†ç¦»åº¦")

OPTIMAL_K = int(optimal_k)
```

### 7.2.8 èšç±»ç»“æœç‰¹å¾åˆ†æ


```python
# åˆ†ææ¯ä¸ªèšç±»ç¾¤ç»„çš„ç‰¹å¾

# é€‰æ‹©å…³é”®ä¸šåŠ¡æŒ‡æ ‡è¿›è¡Œåˆ†æ
key_features = [
    'Age', 'Income',
    'Last_Login_Days_Ago', 'Purchase_Frequency', 'Total_Spending',
    'Average_Order_Value', 'Time_Spent_on_Site_Minutes', 'Pages_Viewed'
]

cluster_profiles = df_clustered.groupby('Cluster')[key_features].mean().round(2)

print("="*70)
print("ğŸ“Š å„èšç±»ç¾¤ç»„ç‰¹å¾ç”»åƒ")
print("="*70)
display(cluster_profiles)

# è®¡ç®—æ¯ä¸ªç¾¤ç»„ç›¸å¯¹äºæ•´ä½“çš„ç‰¹å¾åç¦»åº¦
overall_mean = df[key_features].mean()
deviation_matrix = ((cluster_profiles - overall_mean) / overall_mean * 100).round(1)

print("\n" + "="*70)
print("ğŸ“ˆ å„ç¾¤ç»„ç›¸å¯¹æ•´ä½“å¹³å‡å€¼çš„åç¦»åº¦ï¼ˆ%ï¼‰")
print("="*70)
print("ï¼ˆæ­£å€¼è¡¨ç¤ºé«˜äºå¹³å‡ï¼Œè´Ÿå€¼è¡¨ç¤ºä½äºå¹³å‡ï¼‰")
display(deviation_matrix)

# ä¸ºæ¯ä¸ªèšç±»ç¾¤ç»„ç”Ÿæˆä¸šåŠ¡æ ‡ç­¾
def generate_cluster_label(cluster_id, profile, deviation):
    """åŸºäºç‰¹å¾è‡ªåŠ¨ç”Ÿæˆèšç±»æ ‡ç­¾"""
    labels = []
    
    # å¹´é¾„ç‰¹å¾
    if deviation.loc[cluster_id, 'Age'] > 10:
        labels.append("é«˜é¾„")
    elif deviation.loc[cluster_id, 'Age'] < -10:
        labels.append("å¹´è½»")
    
    # æ”¶å…¥ç‰¹å¾
    if deviation.loc[cluster_id, 'Income'] > 15:
        labels.append("é«˜æ”¶å…¥")
    elif deviation.loc[cluster_id, 'Income'] < -15:
        labels.append("ä½æ”¶å…¥")
    
    # æ´»è·ƒåº¦
    if deviation.loc[cluster_id, 'Last_Login_Days_Ago'] < -20:
        labels.append("é«˜æ´»è·ƒ")
    elif deviation.loc[cluster_id, 'Last_Login_Days_Ago'] > 20:
        labels.append("ä½æ´»è·ƒ")
    
    # è´­ä¹°é¢‘ç‡
    if deviation.loc[cluster_id, 'Purchase_Frequency'] > 20:
        labels.append("é«˜é¢‘è´­ä¹°")
    elif deviation.loc[cluster_id, 'Purchase_Frequency'] < -20:
        labels.append("ä½é¢‘è´­ä¹°")
    
    # æ¶ˆè´¹é‡‘é¢
    if deviation.loc[cluster_id, 'Total_Spending'] > 20:
        labels.append("é«˜æ¶ˆè´¹")
    elif deviation.loc[cluster_id, 'Total_Spending'] < -20:
        labels.append("ä½æ¶ˆè´¹")
    
    return " + ".join(labels) if labels else "æ™®é€šç”¨æˆ·"

cluster_business_labels = {}
print("\n" + "="*70)
print("ğŸ·ï¸ è‡ªåŠ¨ç”Ÿæˆçš„ç¾¤ç»„æ ‡ç­¾")
print("="*70)
for cluster_id in cluster_profiles.index:
    label = generate_cluster_label(cluster_id, cluster_profiles, deviation_matrix)
    cluster_business_labels[cluster_id] = label
    print(f"ç¾¤ç»„ {cluster_id}: {label}")
    print(f"  äººæ•°: {cluster_dist[cluster_id]} ({cluster_dist[cluster_id]/len(df)*100:.1f}%)")
    print()
```

```python
# ä¸ºæ¯ä¸ªç¾¤ç»„ç”Ÿæˆè¯¦ç»†çš„è¥é”€å»ºè®®
def generate_marketing_strategy(cluster_id, profile, deviation, label):
    """ç”Ÿæˆè¥é”€ç­–ç•¥å»ºè®®"""
    
    strategies = {
        'æ ¸å¿ƒä»·å€¼': '',
        'ç”¨æˆ·ç‰¹å¾': '',
        'è¥é”€ç­–ç•¥': [],
        'å…³é”®æŒ‡æ ‡': {},
        'ä¼˜å…ˆçº§': ''
    }
    
    # åŸºäºåç¦»åº¦åˆ¤æ–­æ ¸å¿ƒç‰¹å¾
    high_features = []
    low_features = []
    
    for feature in deviation.columns:
        if deviation.loc[cluster_id, feature] > 20:
            high_features.append(feature)
        elif deviation.loc[cluster_id, feature] < -20:
            low_features.append(feature)
    
    # åˆ¤æ–­ç”¨æˆ·ä»·å€¼
    spending_dev = deviation.loc[cluster_id, 'Total_Spending']
    freq_dev = deviation.loc[cluster_id, 'Purchase_Frequency']
    activity_dev = -deviation.loc[cluster_id, 'Last_Login_Days_Ago']  # è´Ÿå·å› ä¸ºå¤©æ•°è¶Šå°è¶Šæ´»è·ƒ
    
    # è®¾å®šä¼˜å…ˆçº§
    if spending_dev > 20 and freq_dev > 20:
        strategies['ä¼˜å…ˆçº§'] = 'â­â­â­â­â­ æœ€é«˜ä¼˜å…ˆçº§'
        strategies['æ ¸å¿ƒä»·å€¼'] = 'é«˜ä»·å€¼VIPå®¢æˆ·'
    elif spending_dev > 20:
        strategies['ä¼˜å…ˆçº§'] = 'â­â­â­â­ é«˜ä¼˜å…ˆçº§'
        strategies['æ ¸å¿ƒä»·å€¼'] = 'é«˜æ¶ˆè´¹æ½œåŠ›å®¢æˆ·'
    elif freq_dev > 20 and activity_dev > 20:
        strategies['ä¼˜å…ˆçº§'] = 'â­â­â­ ä¸­é«˜ä¼˜å…ˆçº§'
        strategies['æ ¸å¿ƒä»·å€¼'] = 'æ´»è·ƒå¿ è¯šå®¢æˆ·'
    else:
        strategies['ä¼˜å…ˆçº§'] = 'â­â­ æ ‡å‡†ä¼˜å…ˆçº§'
        strategies['æ ¸å¿ƒä»·å€¼'] = 'æ™®é€šå®¢æˆ·ç¾¤ä½“'
    
    # ç”Ÿæˆè¥é”€ç­–ç•¥
    if spending_dev > 20:
        strategies['è¥é”€ç­–ç•¥'].append('ğŸ¯ æä¾›VIPä¸“å±æœåŠ¡å’Œé«˜ç«¯äº§å“æ¨è')
    if freq_dev > 20:
        strategies['è¥é”€ç­–ç•¥'].append('ğŸ’ è®¾è®¡ä¼šå‘˜å¿ è¯šåº¦è®¡åˆ’ï¼Œæä¾›ç§¯åˆ†å¥–åŠ±')
    if activity_dev > 20:
        strategies['è¥é”€ç­–ç•¥'].append('ğŸ“± åŠ å¼ºäº’åŠ¨ï¼Œæ¨é€ä¸ªæ€§åŒ–å†…å®¹')
    if activity_dev < -20:
        strategies['è¥é”€ç­–ç•¥'].append('ğŸ“§ å‘é€å¬å›é‚®ä»¶ï¼Œæä¾›ä¸“å±ä¼˜æƒ åˆ¸')
    if spending_dev < -20:
        strategies['è¥é”€ç­–ç•¥'].append('ğŸ’° æ¨é€ä¼˜æƒ æ´»åŠ¨ï¼Œé™ä½è´­ä¹°é—¨æ§›')
    if freq_dev < -20:
        strategies['è¥é”€ç­–ç•¥'].append('ğŸ”„ ä¼˜åŒ–é¦–å•ä½“éªŒï¼Œä¿ƒè¿›å¤è´­')
    
    if not strategies['è¥é”€ç­–ç•¥']:
        strategies['è¥é”€ç­–ç•¥'].append('ğŸ“Š ç»§ç»­ç»´æŠ¤ç°æœ‰æœåŠ¡æ°´å¹³')
    
    # å…³é”®æŒ‡æ ‡
    strategies['å…³é”®æŒ‡æ ‡'] = {
        'å¹³å‡å¹´é¾„': f"{profile.loc[cluster_id, 'Age']:.0f}å²",
        'å¹³å‡æ”¶å…¥': f"Â¥{profile.loc[cluster_id, 'Income']:,.0f}",
        'è´­ä¹°é¢‘ç‡': f"{profile.loc[cluster_id, 'Purchase_Frequency']:.1f}æ¬¡",
        'æ€»æ¶ˆè´¹': f"Â¥{profile.loc[cluster_id, 'Total_Spending']:,.0f}",
        'å¹³å‡å®¢å•ä»·': f"Â¥{profile.loc[cluster_id, 'Average_Order_Value']:.0f}",
    }
    
    return strategies

# ç”Ÿæˆå¹¶å±•ç¤ºè¥é”€ç­–ç•¥
print("="*70)
print("ğŸ“Š K-Meansèšç±»è¥é”€ç­–ç•¥å»ºè®®")
print("="*70)

for cluster_id in sorted(cluster_profiles.index):
    label = cluster_business_labels[cluster_id]
    count = cluster_dist[cluster_id]
    percentage = count / len(df) * 100
    
    strategy = generate_marketing_strategy(cluster_id, cluster_profiles, deviation_matrix, label)
    
    print(f"\n{'='*70}")
    print(f"ã€ç¾¤ç»„ {cluster_id}ã€‘{label}")
    print(f"{'='*70}")
    print(f"ğŸ“Š ç”¨æˆ·è§„æ¨¡: {count}äºº ({percentage:.1f}%)")
    print(f"ğŸ’ æ ¸å¿ƒä»·å€¼: {strategy['æ ¸å¿ƒä»·å€¼']}")
    print(f"â­ {strategy['ä¼˜å…ˆçº§']}")
    
    print(f"\nğŸ“ˆ å…³é”®æŒ‡æ ‡:")
    for key, value in strategy['å…³é”®æŒ‡æ ‡'].items():
        print(f"  â€¢ {key}: {value}")
    
    print(f"\nğŸ¯ è¥é”€ç­–ç•¥:")
    for i, s in enumerate(strategy['è¥é”€ç­–ç•¥'], 1):
        print(f"  {i}. {s}")

print("\n" + "="*70)
print("âœ… è¥é”€ç­–ç•¥ç”Ÿæˆå®Œæ¯•ï¼")
print("="*70)
```
